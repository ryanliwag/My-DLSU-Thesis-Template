```python
def chunks(l, n):
    """Yield successive n-sized chunks from l."""
    for i in range(0, len(l), n):
        yield l[i:i + n]
```

```python
def chunks1(l, n):
    n = max(1, n)
    return (l[i:i+n] for i in range(0, len(l), n))
```
